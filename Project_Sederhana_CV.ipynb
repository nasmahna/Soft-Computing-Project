membuat fungsi untuk menjadikan dataset kita kedalam bentuk dataframe

```python
def proses_img_df(filepaths):
  labels = [str(filepaths[i]).split("/")[-2]
            for i in range(len(filepaths))]
  filepath = pd.Series(filepaths, name='FilePath').astype(str)
  labels = pd.Series(labels, name='Label')

  df = pd.concat([filepath, labels], axis=1)
  df = df.sample(frac=1).reset_index(drop = True)

  return df
```

## Persiapan Data

```python
train_datagen = ImageDataGenerator(
                    rescale=1./255)
                    # rotation_range=20,
                    # horizontal_flip=True,
                    # shear_range = 0.0,
                    # fill_mode = 'nearest')

validation_datagen = ImageDataGenerator(
                    rescale=1./255)
```

```python
image_path = '/content/dataset_final/train/Anggur/001_da507d16.jpg'
```

```python
img = tf.keras.preprocessing.image.load_img(image_path, target_size= (500,500))
img_tensor = tf.keras.preprocessing.image.img_to_array(img)
img_tensor = np.expand_dims(img_tensor, axis=0)
#Uses ImageDataGenerator to flip the images
datagen = ImageDataGenerator(width_shift_range=[-100, 100],
                    height_shift_range=[-100, 100],
                    rotation_range=120, brightness_range=[0.2, 1.5],
                    zoom_range = [0.3, 1.5] ,shear_range=50)
#Creates our batch of one image
pic = datagen.flow(img_tensor, batch_size =1)
plt.figure(figsize=(16, 16))
#Plots our figures
for i in range(1,17):
   plt.subplot(4, 4, i)
   batch = pic.next()
   image_ = batch[0].astype('uint8')
   plt.imshow(image_)
plt.show()
```

```python
train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(120,120),
        batch_size=5,
        class_mode='categorical',
        shuffle=True)


validation_generator = validation_datagen.flow_from_directory(
        val_dir,
        target_size=(120,120),
        batch_size=5,
        class_mode='categorical',
        shuffle=False)
```

```python
model = Sequential([
    Conv2D(128, (3,3), activation='relu', input_shape=(120,120, 3)),
    MaxPooling2D(2, 2),
    BatchNormalization(),
    Dropout(0.10),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2, 2),
    BatchNormalization(),
    Dropout(0.10),
    Conv2D(32, (3,3), activation='relu'),
    MaxPooling2D(2, 2),
    BatchNormalization(),
    Dropout(0.10),
    Flatten(),
    Dense(256, activation='relu'),
    Dense(5, activation='softmax')
])
```

```python
from tensorflow import keras
import numpy as np

img_path = "/content/dataset_final/train/Anggur/001_da507d16.jpg"

def get_img_array(img_path, target_size):
    img = keras.utils.load_img(
        img_path, target_size=target_size)
    array = keras.utils.img_to_array(img)
    array = np.expand_dims(array, axis=0)
    return array

img_tensor = get_img_array(img_path, target_size=(120,120))
```

```python
from tensorflow import keras
from tensorflow.keras import layers

layer_outputs = []
layer_names = []
for layer in model.layers:
    if isinstance(layer, (layers.Conv2D, layers.MaxPooling2D)):
        layer_outputs.append(layer.output)
        layer_names.append(layer.name)
activation_model = keras.Model(inputs=model.input, outputs=layer_outputs)
```

```python
activations = activation_model.predict(img_tensor)
```

```python
first_layer_activation = activations[0]
print(first_layer_activation.shape)
```

```python
images_per_row = 16
for layer_name, layer_activation in zip(layer_names, activations):
    n_features = layer_activation.shape[-1]
    size = layer_activation.shape[1]
    n_cols = n_features // images_per_row
    display_grid = np.zeros(((size + 1) * n_cols - 1,
                             images_per_row * (size + 1) - 1))
    for col in range(n_cols):
        for row in range(images_per_row):
            channel_index = col * images_per_row + row
            channel_image = layer_activation[0, :, :, channel_index].copy()
            if channel_image.sum() != 0:
                channel_image -= channel_image.mean()
                channel_image /= channel_image.std()
                channel_image *= 64
                channel_image += 128
            channel_image = np.clip(channel_image, 0, 255).astype("uint8")
            display_grid[
                col * (size + 1): (col + 1) * size + col,
                row * (size + 1) : (row + 1) * size + row] = channel_image
    scale = 1. / size
    plt.figure(figsize=(scale * display_grid.shape[1],
                        scale * display_grid.shape[0]))
    plt.title(layer_name)
    plt.grid(False)
    plt.axis("off")
    plt.imshow(display_grid, aspect="auto", cmap="viridis")
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=15,
    verbose=1
)
```

## VGG16

image.png

```python
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
from tf.keras.applications.mobilenet import MobileNet
```

```python
base_vgg = VGG16(
    include_top=False,
    weights='imagenet',
    input_shape=(120,120,3),
    pooling='avg'
)
```

```python
base_vgg.trainable=False
```

```python
inputs = base_vgg.input

x = tf.keras.layers.Dense(128, activation='relu')(base_vgg.output)
x = tf.keras.layers.Dense(256, activation='relu')(x)
x = tf.keras.layers.Dropout(0.10)(x)

outputs = tf.keras.layers.Dense(5, activation='softmax')(x)

model_vgg = tf.keras.Model(inputs=inputs, outputs=outputs)
```

```python
model_vgg.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history_vgg = model_vgg.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=15,
    verbose=1
)
```

```python
from matplotlib import pyplot
# plot learning curves
pyplot.title('Learning Curves')
pyplot.xlabel('Epoch')
pyplot.ylabel('Loss function')
pyplot.plot(history_vgg.history['loss'], label='train')
pyplot.plot(history_vgg.history['val_loss'], label='val')
pyplot.legend()
pyplot.show()
```

```python
pyplot.title('Learning Curves')
pyplot.xlabel('Epoch')
pyplot.ylabel('Accuracy')
pyplot.plot(history_vgg.history['accuracy'], label='train')
pyplot.plot(history_vgg.history['val_accuracy'], label='val')
pyplot.legend()
pyplot.show()
```
